```python
def MC(G,T, r = 0.): 
    """Monte Carlo price of the reward G (array). 
       T = maturity, r = continuous interest rate. """
    return np.mean(np.exp(-r*T) * G,axis = -1)
```

### 6.1.1 Asian call option

$$    g(X) = \left(\bar{a}(X) - K \right)^+, \quad  \bar{a}(X) = \frac{1}{T}\int_0^t X_udu,  \quad X \in \Lambda_u. $$

```python
x0      = 100     # Initial value
K,T     = x0,1/4  # Option parameters 
t       = T/3     # Today's date
r,sig   = 0.0,0.2 # Black-Scholes Parameters (no dividends)
# Number of time steps
# 1-hour time increment (assuming 252 business days/year and 24 hour trading sessions)
dt  = 1/(252*24)          
ts  = np.arange(0,T+dt,dt)  # Time grid (flat)
Ts  = ts[:,__]              # Time grid (column vector)

# 1-day time increment (assuming 252 business days/year and 24 hour trading sessions)
# dt  = 1/(252)          
# ts  = np.arange(0,T+dt,dt)  # Time grid (flat)
# Ts  = ts[:,__]              # Time grid (column vector)

# It√¥ map linking Brownian motion to the coordinate process Y 
# under Q (here Black-Scholes)
w2Y = lambda x,Ts,W: x * np.exp(sig * W + (r - sig**2/2) * Ts)

# Payoff
g = lambda X: np.maximum(np.mean(X,axis = 0) - K,0.)
```

```python
# Today's date (time step)
n = int(t/dt)
# Observed path 
X = w2Y(x0,Ts[:n+1],BM(Ts[:n+1],1))
# Shock
J = 10; Y = w2Y(X[-1],Ts[n:]-t,BM(Ts[n:]-t,J)) 
# Path 
Z = np.vstack([np.tile(X,(1,J)),Y])
#plt.plot(Ts,Z,"--",color = "k",zorder = 3)
plt.plot(Ts[:n+1],X)
plt.plot(Ts[n:],Y)
plt.show()
```

```python
def f(t,X,J = int(1e4)):
    """Price functional f(X) = E[g(Z)|X]. 
       J = # Monte Carlo simulations"""
    # Shocks
    Ts_ = Ts[int(t/dt):]-t # time grid
    Y   = w2Y(X[-1,:],Ts_,BM(Ts_,J)) 
    # Concatenated paths
    Z   = np.vstack([np.tile(X,(1,J)),Y])
    # Monte Carlo price
    return MC(g(Z),T)
```

```python
# Today's date (time step)
n = int(t/dt)
# Observed path 
X = w2Y(x0,Ts[:n+1],BM(Ts[:n+1],1))
X[-1]
# Today's price
f_ = f(t,X)
```

```python
tic()
# Functional derivatives (Greeks). Use finite difference
dx        = x0*1e-2
dx_       = dx * (Ts[:n+1] == t) # bump (column vector)
fUp,fDown = f(t,X + dx_),f(t,X - dx_) # bumped path at final time (up an down)
# Distorted paths for the Lie derivative
X_xt      = np.vstack([X,X[-1,:]]) + dx * (Ts[:n+2] == (t+dt)) 
X_tx      = np.vstack([X + dx_,X[-1,:]+ dx])

lbda = {"t"    : (f(t+dt,np.vstack([X,X[-1]])) - f_)/dt,
        "x"    : (fUp - f_)/dx,
        "xx"   : (fUp - 2*f_ + fDown)/dx**2, # butterfly approximation
        "[t,x]": (f(t+dt,X_xt) - f(t+dt,X_tx))/(dt*dx) } # Lie derivative
print("Approximated PnL - Run time (offline): %2.4f sec."%toc())
lbda
```

```python
X_xtUp    = np.vstack([X + dx_,X[-1,:] + dx]) 
X_txUp    = np.vstack([X,X[-1,:]]) + dx * (Ts[:n+2] == (t+dt))
X_xtDown  = np.vstack([X - dx_,X[-1,:] - dx]) 
X_txDown  = np.vstack([X,X[-1,:]]) - dx * (Ts[:n+2] == (t+dt))

for jj in range(30): 
    L1 = (f(t+dt,X_xtUp) - f(t+dt,X_txUp))/(dt*dx)
    
    # Test: (x centered) Lie derivative 
    L2 = (f(t+dt,X_xtUp) - f(t+dt,X_xtDown) \
       -  f(t+dt,X_txUp) + f(t+dt,X_txDown) )/(2*dt*dx)
    plt.scatter([L1],[L2])
plt.xlabel("Forward"); plt.ylabel("Centered")
#print(L1,L2)
```

Fast computation of PnL
$$(f,X) \; \overset{\text{offline}}{\longrightarrow} \;
  (\Theta,\Delta,\Gamma,\mathcal{L}) \; \overset{\text{online}}{\longrightarrow} \; 
       \text{P&L}(Y) =   \Theta \delta t + \Delta \delta y + \Gamma\frac{\delta y ^2}{2} + \mathcal{L} \mathcal{A}(Y).$$


```python
def PnL(deltaT,deltaY,A,lbda):
    """Taylor approximation of the P&L f(X*Y) - f(X) using the Greeks
       (Theta,Delta,Gamma,L)."""
    return lbda["t"] * deltaT + lbda["x"] * deltaY + lbda["xx"] * deltaY**2/2 \
         + lbda["[t,x]"] * A

# One-day schock paths
deltaT  = 1/252                      # Time increment
N_      = 24                         # one quote per hour         
ts_     = np.linspace(0,deltaT,N_+1) # Time grid (flat)
Ts_     = ts_[:,__]                  # Time grid (column vector) 

# One-week schock paths
# deltaT  = 5/252                      # Time increment
# N_      = 5                          # five business days per week         
# ts_     = np.linspace(0,deltaT,N_+1) # Time grid (flat)
# Ts_     = ts_[:,__]                  # Time grid (column vector) 
# Integral weights for trapezoidal rule (or use scipy.integrate)
w = 2**(-np.sum(np.eye(N_+1)[:,[0,-1]],axis = 1,keepdims = 1))

# Future scenarios
J = 50; Y = w2Y(X[-1],Ts_,BM(Ts_,J))
# Exact PnL
tic()
PnLEx = [f(t+deltaT,np.vstack([X,Y[:,[j]]])) - f_ for j in range(J)]
print("Exact PnL - Run time: %2.4f sec."%toc())
#============= Approximated PnL using the FTE ============#
# Approximated PnL
tic()
# Space increments
deltaY = Y[-1] - Y[0]
# L√©vy area
A = np.sum((Y - Y[0]) * w * dt,axis = 0) - deltaT*deltaY/2
PnLAp = PnL(deltaT,deltaY,A,lbda)
print("Approximated PnL - Run time (online): %2.4f sec."%toc())
```

### Fast computation of VaR

```python
# Confidence levels
alphas = np.arange(0.01,0.11,0.01)#[0.01,0.05,0.1]
VaREx  = np.quantile(PnLEx,alphas)
VaRAp  = np.quantile(PnLAp,alphas)

save   = 0
fig,ax = plt.subplots(1,1,figsize = (5.5,5))
ax.set_facecolor("aliceblue");ax.set_prop_cycle(colCyc)
ax.plot(alphas,VaREx,"--o",label = "Exact")
ax.plot(alphas,VaRAp,"-o" ,label = "Approximated")
ax.set_xticks(alphas,["%2.f"%(100*a) + "\%" for a in alphas])
ax.grid(color = "midnightblue",alpha= 0.1)
ax.set_xlabel("Confidence level",fontsize = 13)
ax.set_ylabel("VaR",rotation = 0,fontsize = 13,labelpad = 13)
plt.legend(frameon = False,fontsize = 15)
if save: plt.savefig("Figures/VaR.pdf",dpi = 500,bbox_inches ="tight",pad_inches = 0.15)
plt.show()
```

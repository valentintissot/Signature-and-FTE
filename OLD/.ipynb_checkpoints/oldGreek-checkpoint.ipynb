### 6.1.1 Asian call option

$$    g(X) = \left(\bar{a}(X) - K \right)^+, \quad  \bar{a}(X) = \frac{1}{T}\int_0^t X_udu,  \quad X \in \Lambda_u. $$

```python
x0      = 100     # Initial value
K,T     = x0,1/2  # Option parameters 
t       = T/3     # Today's date
r,sig   = 0.0,0.2 # Black-Scholes Parameters (no dividends)

# 1-day time increment (assuming 252 business days/year and 24 hour trading sessions)
dt  = 1/252          
ts  = np.arange(0,T+dt,dt)  # Time grid (flat)
Ts  = ts[:,__]              # Time grid (column vector)
N,n = int(T/dt), int(t/dt)

# It√¥ map linking Brownian motion to the coordinate process Y 
# under Q (here Black-Scholes)
w2Y = lambda x,Ts,W: x * np.exp(sig * W + (r - sig**2/2) * Ts)

# Payoff of 100 Asian call options
g = lambda X: 100 * np.maximum(np.mean(X,axis = 0) - K,0.)
```

```python
# Presimulated Brownian paths for the price functional
W = BM(Ts[n:] - t,int(2**16))

def f(t,X):
    """Price functional f(X) = E[g(Z)|X]. 
       J = # Monte Carlo simulations"""
    n_  = int(t/dt)
    Ts_ = Ts[n_:]-t # time grid
    # Shocks
    Y   = w2Y(X[-1],Ts_,W[:N-n_+1])  
    # Concatenated paths
    Z   = np.vstack([np.tile(X,(1,np.shape(Y)[1])),Y])
    # Monte Carlo price
    return MC(g(Z),T)
```

```python
# Observed path 
rdm.seed(1)
X = w2Y(x0,Ts[:n+1],BM(Ts[:n+1],1)) 
# Today's option price
f_ = f(t,X)

Y = w2Y(X[-1],Ts[n:]-t,BM(Ts[n:]-t,200)) 

fig,ax = plt.subplots(1,1,figsize = (5.5,4))
ax.set_facecolor("aliceblue"); ax.set_prop_cycle(colCyc)
ax.plot(Ts[:n+1],X)
plt.plot(Ts[n:],Y,alpha = 0.5)
plt.xticks([0,t,T]); plt.grid(color = cols[0], alpha = 0.5, axis = "x")
plt.show()
print(X[-1],g(X))
```

```python
tic()
# Functional derivatives (Greeks). Use finite difference
dx        = x0 * np.sqrt(dt)*1e-1
dx_       = dx * (Ts[:n+1] == t) # bump (column vector)
fUp,fDown = f(t,X + dx_),f(t,X - dx_) # bumped path at final time (up an down)
# Distorted paths for the Lie derivative
X_xtUp    = np.vstack([X + dx_,X[-1,:] + dx]) 
X_txUp    = np.vstack([X,X[-1,:]]) + dx * (Ts[:n+2] == (t+dt))
X_xtDown  = np.vstack([X - dx_,X[-1,:] - dx]) 
X_txDown  = np.vstack([X,X[-1,:]]) - dx * (Ts[:n+2] == (t+dt))

lbda = {"t"    : (f(t+dt,np.vstack([X,X[-1]])) - f_)/dt,
        "x"    : (fUp - fDown)/(2*dx),                   # centered difference 
        "xx"   : (fUp - 2*f_ + fDown)/dx**2,             # butterfly approximation
        "[t,x]": (f(t+dt,X_xtUp) - f(t+dt,X_xtDown) \
               -  f(t+dt,X_txUp) + f(t+dt,X_txDown) )/(2*dt*dx) } # Lie derivative

print("Greeks calculation - Run time: %2.2f sec.\n"%toc())

print("Portfolio value: %2.2f\n"%f_)
display(pd.DataFrame.from_dict(lbda,orient='index',columns = ["Greeks"]).round(2))
```

Fast computation of PnL
$$(f,X) \; \overset{\text{offline}}{\longrightarrow} \;
  (\Theta,\Delta,\Gamma,\mathcal{L}) \; \overset{\text{online}}{\longrightarrow} \; 
       \text{P&L}(Y) =   \Theta \delta t + \Delta \delta y + \Gamma\frac{\delta y ^2}{2} + \mathcal{L} \mathcal{A}(Y).$$


```python
def PnL(deltaT,deltaY,A,lbda):
    """Taylor approximation of the P&L f(X*Y) - f(X) using the Greeks
       (Theta,Delta,Gamma,L)."""
    return lbda["t"] * deltaT + lbda["x"] * deltaY + lbda["xx"] * deltaY**2/2 \
         + lbda["[t,x]"] * A

# One-week shock paths
N_      = 5                          # five business days per week   
deltaT  = N_/252                     # time increment
ts_     = np.linspace(0,deltaT,N_+1) # Time grid (flat)
Ts_     = ts_[:,__]                  # Time grid (column vector) 

# Integral weights for trapezoidal rule (or use scipy.integrate)
w = 2**(-np.sum(np.eye(N_+1)[:,[0,-1]],axis = 1,keepdims = 1))

# Future scenarios
J = 5000; Y = w2Y(X[-1],Ts_,BM(Ts_,J))
# Exact PnL
tic()
PnLEx = [f(t+deltaT,np.vstack([X,Y[:,[j]]])) - f_ for j in range(J)]
print("Exact PnL - Run time: %2.2f sec."%toc())
#============= Approximated PnL using the FTE ============#
# Approximated PnL
tic()
# Space increments
deltaY = Y[-1] - Y[0]
# L√©vy area
A = np.sum((Y - Y[0]) * w * dt,axis = 0) - deltaT*deltaY/2
PnLAp = PnL(deltaT,deltaY,A,lbda)
print("Approximated PnL - Run time (online): %2.2f sec."%toc())
```

```python
save   = 0
fig,ax = plt.subplots(1,1,figsize = (7,5))
ax.set_facecolor("aliceblue")
ax.set_prop_cycle(colCyc)
xMin,xMax = np.quantile([PnLEx,PnLAp],[0.,1.])
plt.hist(PnLEx,bins = 35,alpha = 0.4,density = True,label = "Exact")#,,range = [xMin,xMax]
plt.hist(PnLAp,bins = 35,alpha = 0.4,density = True,label = "FTE")#,range = [xMin,xMax]
plt.legend(frameon = False,fontsize = 15)
if save: plt.savefig("Figures/histPnL.pdf",dpi = 500,bbox_inches ="tight",pad_inches = 0.15)
plt.show()
```

### Fast computation of Risk Measures
#### 1. Value-at-Risk (VaR)

```python
# Confidence levels
alphas = np.arange(0.01,0.055,0.005)#[0.01,0.05,0.1]
VaREx  = np.quantile(PnLEx,alphas)
VaRAp  = np.quantile(PnLAp,alphas)

save   = 1
fig,ax = plt.subplots(1,1,figsize = (5.5,5))
ax.set_facecolor("aliceblue");ax.set_prop_cycle(colCyc)
ax.plot(alphas,VaREx,"--o",label = "Exact")
ax.plot(alphas,VaRAp,"-o" ,label = "FTE")
ax.set_xticks(alphas,["%2.1f"%(100*a) + "\%" for a in alphas])
ax.grid(color = "midnightblue",alpha= 0.1,axis = "y")
ax.set_xlabel("Level",fontsize = 13)
ax.set_ylabel("VaR",rotation = 0,fontsize = 13,labelpad = 13)
plt.legend(frameon = False,fontsize = 15,loc = "lower center",ncol = 2)
ax.set_ylim(np.min([VaREx,VaRAp])*1.05,np.max([VaREx,VaRAp])*0.95)
if save: plt.savefig("Figures/VaR.pdf",dpi = 500,bbox_inches ="tight",pad_inches = 0.15)
plt.show()
```

#### 2. Expected Shortfall (ES)

```python
def ES(PnL,VaR,alpha): 
    """Compute expected shortfall at level alpha (with associated VaR)
       of a PnL array."""
    return np.mean(np.where(PnL <= VaR,PnL,0.))/alpha

ESEx  = [ES(PnLEx,v,a) for v,a in zip(VaREx,alphas)]
ESAp  = [ES(PnLAp,v,a) for v,a in zip(VaRAp,alphas)]


cols = ["midnightblue","darkred",'purple','indianred',"plum","gold","steelblue"]
colCyc = cyc('color', cols) #linestyle = ["-","--",":"])

save   = 1
fig,ax = plt.subplots(1,1,figsize = (6,4.5))
ax.set_facecolor("aliceblue"); ax.set_prop_cycle(colCyc)
ax.plot(alphas,VaREx,":o",label = "VaR, exact",markersize = 5)
ax.plot(alphas,VaRAp,"-.o" ,label = "VaR, FTE",markersize = 5)
ax.plot(alphas,ESEx,"-o",label = "ES, exact",markersize = 5)
ax.plot(alphas,ESAp,"--o" ,label = "ES, FTE",markersize = 5)
ax.set_xticks(alphas,["%2.1f"%(100*a) + "\%" for a in alphas])
ax.grid(color = "midnightblue",alpha= 0.1,axis = "y")
ax.set_xlabel("Level",fontsize = 13)
#ax.set_ylabel("ES",rotation = 0,fontsize = 13,labelpad = 13)
plt.legend(frameon = False,fontsize = 12)
#ax.set_ylim(np.min([ESEx,ESAp])*1.05,np.max([ESEx,ESAp])*0.9)
ax.set_ylim(np.min([ESEx,ESAp])*1.05,np.max([VaREx,VaRAp])*0.975)
if save: plt.savefig("Figures/VaRES.pdf",dpi = 500,bbox_inches ="tight",pad_inches = 0.15)
plt.show()
```
